```python
# line plot of time series
from pandas import Series
from matplotlib import pyplot
print ("libraries loaded.")
```

```python

# load dataset
series = Series.from_csv('monthly-car-sales-in-quebec-1960.csv', header=0)
print (series.shape)
# display first few rows
print(series.head(5))
# line plot of dataset
series.plot()
pyplot.show()
```

```python
print (type(series))
print (series.head())
# seasonally adjust the time series
# seasonal difference
differenced = series.diff(12)
#https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.diff.html

print (type(differenced))
print (differenced.head())
# trim off the first year of empty data
differenced = differenced[12:]
# save differenced dataset to file
differenced.to_csv('seasonally_adjusted.csv')
# plot differenced dataset
differenced.plot()
pyplot.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(differenced)
pyplot.show()
```

```python
from pandas import Series
from statsmodels.graphics.tsaplots import plot_acf
from matplotlib import pyplot
series = Series.from_csv('seasonally_adjusted.csv', header=None)
plot_acf(series)
pyplot.show()
```
